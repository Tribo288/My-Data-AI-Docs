# Hàm F1-Score thường được dùng để đánh giá mô hình phân loại
# F1-Score function used to evaluate sorting model
Precision = TP/(TP +FP)
Recall =  TP/(TP+FN)
F1-Score = 2*(Precision* Recall)/(Precision +Recall)

```python
# check number
def is_number(n):
    try:
        float(n)
        return True
    except ValueError:
        return False

if "__name__" == "__main__":
    while True :
        tp,fp,fn = list(map(int, input().split()))
        ls = [tp,fp,fn]
        for i in ls:
            if is_number(i) == False:
                print(f"{str(ls[i])} must be int")
        precision = tp/(tp+fp)
        recall = tp/(tp+fn)
        F1_score = 2*((precision*recall)/(precision+recall))
        out = [precision,recall,F1_score]
        # in kq
        for i in out :
            print(f"{str(i)} = {i}") 
```
