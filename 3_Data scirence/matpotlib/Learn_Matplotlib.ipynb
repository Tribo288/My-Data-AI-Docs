```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
```

# Scatter plot (phân tán)

```python
plt.scatter(x, y)
plt.show()
```

# Bar chart (cột)

```python
names = ["A", "B", "C"]
scores = [8, 7, 9]

plt.bar(names, scores)
plt.show()
```

# Histogram (phân phối)

```python
data = [1,2,2,3,3,3,4,4,5]
plt.hist(data, bins=5)
plt.show()
```

# title,..


```python
x = [1,3,5,6,10,12,15,16,18]
y = [int(i) for i in range(1,10)]


plt.plot(x, y)

plt.title("So sánh độ ngu")
plt.xlabel("Số tuổi")
plt.ylabel("độ ngu")
plt.show()
```
